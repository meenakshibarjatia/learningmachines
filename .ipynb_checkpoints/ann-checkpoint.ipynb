```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original')
X_train = mnist.data[:60000];
y_train = mnist.target[:60000];
X_test = mnist.data[60000:];
y_test = mnist.target[60000:];

#normalize
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.utils import shuffle

stdscl = StandardScaler();
onehot = OneHotEncoder();

X_train_norm = stdscl.fit_transform(X_train)
y_train_norm = onehot.fit_transform(y_train.reshape(-1, 1))
y_train = y_train.astype('int')
y_test = y_test.astype('int')

X_train_norm, y_train = shuffle(X_train_norm, y_train)

plt.imshow(X_train_norm[1].reshape(28, 28), cmap = matplotlib.cm.binary, interpolation = "nearest")
print(y_train[1])
plt.show()
```

```python
print(y_train)
```

```python
feat_col = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units = [300, 100], n_classes = 10, feature_columns = feat_col)
dnn_clf = tf.contrib.learn.SKCompat(dnn_clf)
dnn_clf.fit(x = X_train_norm, y = y_train.reshape(-1, 1), batch_size = 50, steps = 40000)
```

```python
from sklearn.metrics import accuracy_score

y_pred = dnn_clf.predict(X_test)
acc = accuracy_score(y_pred['classes'], y_test)
print(acc)
```
