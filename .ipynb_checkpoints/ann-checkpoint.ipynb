```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original')
X_train = mnist.data[:60000];
y_train = mnist.target[:60000];
X_test = mnist.data[60000:];
y_test = mnist.target[60000:];

#normalize
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler

stdscl = MinMaxScaler(feature_range = (0,1));
onehot = OneHotEncoder();

X_train_norm = stdscl.fit_transform(X_train)
y_train_norm = onehot.fit_transform(y_train.reshape(-1, 1))
X_test_norm = stdscl.transform(X_test)

y_train = y_train.astype('int')
y_test = y_test.astype('int')

X_train_norm, y_train = shuffle(X_train_norm, y_train)

plt.imshow(X_train_norm[1].reshape(28, 28), cmap = matplotlib.cm.binary, interpolation = "nearest")
print(y_train[1])
plt.show()
```

```python
print(y_train)
```

```python
feat_col = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units = [300, 100], n_classes = 10, feature_columns = feat_col)
dnn_clf = tf.contrib.learn.SKCompat(dnn_clf)
dnn_clf.fit(x = X_train_norm, y = y_train.reshape(-1, 1), batch_size = 50, steps = 40000)
```

```python
from sklearn.metrics import accuracy_score

y_pred = dnn_clf.predict(X_test)
acc = accuracy_score(y_pred['classes'], y_test)
print(acc)
```

```python
#train dnn using tf

n_inputs = X_train_norm.shape[1]
n_hidden1 = 100
n_hidden2 = 50
n_output = len(np.unique(y_train))

X = tf.placeholder(tf.float32, shape = (None, n_inputs), name = 'X')
y = tf.placeholder(tf.int64, shape = (None), name = 'y')

def neuron_layer(X, n_neurons, name, activation = None):
    with tf.name_scope(name):
        n_inputs = int(X.get_shape()[1])
        stddev = 2/np.sqrt(n_inputs)
        init = tf.truncated_normal((n_inputs, n_neurons), stddev = stddev)
        W = tf.Variable(init, name = 'weights')
        b = tf.Variable(tf.zeros([n_neurons]), name = 'biases')
        z = tf.matmul(X, W) + b
        if activation == 'relu':
            return tf.nn.relu(z)
        else:
            return z
        
with tf.name_scope('dnn'):
    hidden1 = neuron_layer(X, n_hidden1, 'hidden1', activation = 'relu')
    hidden2 = neuron_layer(hidden1, n_hidden2, 'hidden2', activation = 'relu')
    logits = neuron_layer(hidden2, n_output, 'outputs')
    
with tf.name_scope('loss'):
    xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels = y, logits = logits)
    loss = tf.reduce_mean(xentropy, name = "loss")
    
learning_rate = 0.01

with tf.name_scope('train'):
    optimizer = tf.train.GradientDescentOptimizer(learning_rate)
    training_op = optimizer.minimize(loss)
    
with tf.name_scope('eval'):
    cor = tf.nn.in_top_k(logits, y, 1)
    accuracy = tf.reduce_mean(tf.cast(cor, tf.float32))
    
init = tf.global_variables_initializer()
saver = tf.train.Saver()
```

```python
n_epochs = 100
batch_size = 50
m = X_train_norm.shape[0]

with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        X_trainsh, y_trainsh = shuffle(X_train_norm, y_train)
        for batch in range(int(m/batch_size)):
            xbatch = X_trainsh[batch_size*batch : batch_size*batch + batch_size - 1]
            ybatch = y_trainsh[batch_size*batch : batch_size*batch + batch_size - 1]
            sess.run(training_op, feed_dict = {X : xbatch, y : ybatch})
        acc_train = accuracy.eval(feed_dict = {X : xbatch, y : ybatch})
        acc_test = accuracy.eval(feed_dict = {X : X_test_norm, y : y_test})
        print(epoch, "training err : ", acc_train, "Testing error : ", acc_test)
    save_path = saver.save(sess, './ann_trial_final.ckpt')
            
```
